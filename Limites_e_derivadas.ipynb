### 1.2 Limites Laterais

* * *

Limites laterais referem-se ao comportamento de uma função à medida que a variável independente se aproxima de um ponto específico por um lado (esquerda ou direita). Eles são importantes para entender o comportamento de funções em pontos onde a função pode ter descontinuidades ou onde o limite bilateral não existe.

**Limite Lateral à Direita:**

O limite lateral à direita de uma função $f(x)$ quando $x$ se aproxima de um ponto $a$ é denotado por $\lim_{x \to a^+} f(x)$. Isso significa que estamos considerando os valores de $f(x)$ para valores de $x$ que são maiores que $a$ e se aproximam de $a$.

**Limite Lateral à Esquerda:**

O limite lateral à esquerda de uma função $f(x)$ quando $x$ se aproxima de um ponto $a$ é denotado por $\lim_{x \to a^-} f(x)$. Isso significa que estamos considerando os valores de $f(x)$ para valores de $x$ que são menores que $a$ e se aproximam de $a$.

**Relação entre Limites Laterais e Limite Bilateral:**

O limite bilateral de uma função $f(x)$ quando $x$ se aproxima de um ponto $a$, denotado por $\lim_{x \to a} f(x)$, existe se e somente se ambos os limites laterais (à direita e à esquerda) existem e são iguais. Ou seja:

$\lim_{x \to a} f(x) = L$ se e somente se $\lim_{x \to a^+} f(x) = L$ e $\lim_{x \to a^-} f(x) = L$.

Se os limites laterais forem diferentes ou se um deles não existir, o limite bilateral não existe.

**Exemplo:**

Considere a função definida por partes:

$f(x) = \begin{cases} x + 2 & \text{se } x < 1 \\ x^2 & \text{se } x \ge 1 \end{cases}$

Vamos determinar os limites laterais quando $x$ se aproxima de 1.

Limite lateral à esquerda quando x tende a 1: 3

Limite lateral à direita quando x tende a 1: 1

Os limites laterais são diferentes, portanto o limite bilateral não existe.

**Outro Exemplo:**

Considere a função:

$g(x) = \frac{|x|}{x}$

Vamos determinar os limites laterais quando $x$ se aproxima de 0.

Limite lateral à esquerda quando x tende a 0: -1

Limite lateral à direita quando x tende a 0: 1

Os limites laterais são diferentes, portanto o limite bilateral não existe.
